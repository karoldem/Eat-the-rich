```python
from pandas import read_csv, merge

q3 = read_csv('https://raw.githubusercontent.com/karoldem/Eat-the-rich/main/ilc_di01_page_linear.csv')
avg = read_csv('https://raw.githubusercontent.com/karoldem/Eat-the-rich/main/nama_10_fte__custom_4232263_page_linear.csv')

q3  =  q3.loc [   q3['TIME_PERIOD'] == 2021   ]
q3  =      q3 [   ['geo', 'OBS_VALUE']        ]
avg =     avg [   ['geo', 'OBS_VALUE']        ]

q3 = q3.astype({"geo":'string', "OBS_VALUE":'int64'})
avg = avg.astype({"geo":'string', "OBS_VALUE":'int64'})


df = merge(   q3,   avg,   on='geo',   how='inner',   suffixes=( '_q3', '_avg' )   ).rename(columns={'OBS_VALUE_q3':'q3', 'OBS_VALUE_avg':'avg'})
#df.loc[(df.q3 * df.avg).sort_values().index]
df['avg/q3'] = df.apply(lambda row: row.avg/row.q3, axis = 1)
df.sort_values('avg/q3')
```
